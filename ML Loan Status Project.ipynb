### separate input and output features

```python
x=df1.drop(labels='Loan Status',axis=1)
y=df1[['Loan Status']]
```

### separating categorical and numerical columns from x

```python
cat=[]
con=[]
for i in x.columns:
    if x[i].dtypes==object:
        cat.append(i)
    else:
        con.append(i)
```

```python
xcat = x[cat]
xcon = x[con]
```

### feature encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for i in xcat.columns:
    xcat[i]=le.fit_transform(xcat[i])
```

### feature scaling using standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
xcon=pd.DataFrame(ss.fit_transform(xcon),columns=con)
```

### join xcat and xcon

```python
x=xcon.join(xcat)
```

### Outliers Treatment and reindexing

```python
out=[]
for i in xcon.columns:
    ol=xcon[(xcon[i]<-3)|(xcon[i]>3)].index
    out.extend(ol)
```

```python
outliers=list(set(out))
```

```python
outliers
```

```python
x = x.drop(index=outliers,axis=0)
y = y.drop(index=outliers,axis=0)
```

```python
x.index = range(0,x.shape[0])
y.index = range(0,y.shape[0])
```

```python
y=le.fit_transform(y)
```

### Applying stratified k fold cross validation

```python
xtrain_sm = pd.DataFrame(xtrain_sm)
ytrain_sm = pd.Series(ytrain_sm)
```

```python
from sklearn.model_selection import StratifiedKFold

accuracy=[]

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
skf.get_n_splits(xtrain_sm,ytrain_sm)

for train_index, test_index in skf.split(xtrain_sm, ytrain_sm):
    xtrain_skf, xtest_skf = xtrain_sm.iloc[train_index], xtrain_sm.iloc[test_index]
    ytrain_skf, ytest_skf = ytrain_sm.iloc[train_index], ytrain_sm.iloc[test_index]

    adb.fit(xtrain_skf, ytrain_skf)

    ypred_skf = adb.predict(xtest_skf)

    score = accuracy_score(ypred_skf,ytest_skf)
    accuracy.append(score)

print(accuracy)
```

```python
np.array(accuracy).mean()
```
